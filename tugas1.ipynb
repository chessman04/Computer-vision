```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data (jumlah hari penjualan dan banyak penjualan)
X = np.array([2,4,6,8,9]).reshape(-1, 1)  # jumlah hari penjualan (fitur)
y = np.array([2,6,8,12,4])  # Banyak penjualan (target)

# Membuat model regresi linear
model = LinearRegression()
model.fit(X, y)

# Koefisien dan Intersep
b = model.coef_[0]  # Gradien (slope)
a = model.intercept_  # Intersep (intercept)
print(f"b :  {b:.2f}")
print(f"a :  {a:.2f}")
print(f"Persamaan garis regresi: y = {a:.2f} + {b:.2f}x")

# banyak penjualan pada hari ke 10
prediksi = model.predict(np.array([[10]]))
print(f"banyak penjualan : {prediksi[0]:.2f} ")
```

```python
# Visualisasi
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, model.predict(X), color='red', label='Garis Regresi')
plt.scatter(6, prediksi, color='green', label='Prediksi (Banyak Penjualan Hari Ke-10)')
plt.xlabel('Jumlah Hari Penjualan')
plt.ylabel('Banyak Penjualan')
plt.legend()
plt.title('Regresi Linear: Jumlah Hari & Banyak Penjualan')
plt.show()
```
